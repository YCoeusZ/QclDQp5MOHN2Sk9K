```python
df=pd.read_csv("../data/raw.csv")
df["y"]=df["y"].map({"no": 0, "yes": 1})
features=df.columns[:-1]
df_feat=df[features]
df_tar=df[["y"]]

data_transform_layer=data_processing.data_transform()
data_transform_layer.fit(X=df_feat,y=df_tar)
df_feat_transformed=data_transform_layer.transform(X=df_feat)
df_feat_transformed
```

```python
df_transformed=pd.concat([df_feat_transformed, df_tar], axis=1)
df_transformed
```

```python
# Printing all non-trivial dangerous correlation 
bar=0.7
df_feat_corr=df_feat_transformed.corr()
df_feat_corr_cut=df_feat_corr[df_feat_corr.abs()>=bar]
for feature in df_feat_transformed.columns: 
    cur_df=df_feat_corr_cut[[feature]]
    cur_list=list(cur_df.dropna().index)
    cur_list.remove(feature)
    cur_list=np.array(cur_list)
    cur_corr=cur_df.loc[cur_list, feature].round(2).astype("str").values
    cur_print=cur_list+np.array([" with "]*len(cur_list))+cur_corr
    if len(cur_list): 
        print(f"The features have correlation above {bar} with {feature}: {cur_print}. ")
```

F1 scores 

```python
scaler=StandardScaler()
f_score, p_value= f_classif(X=scaler.fit_transform(X=df_feat_transformed),y=df_tar["y"])
df_f=pd.DataFrame({
                "features": df_feat_transformed.columns, 
                "f score": f_score.round(2), 
                "p value": p_value.round(2)
            })
df_f
```

```python
df_f["rank"]=df_f["f score"].rank(method="dense", ascending=False).astype(int)
df_f
```

```python
df_f_rankded=df_f.sort_values(by="rank",ascending=True).reset_index(drop=True)
df_f_rankded
```

Mutual information

```python
scaler=StandardScaler()
mi_score_arr=np.array([mutual_info_classif(X=scaler.fit_transform(X=df_feat_transformed),y=df_tar["y"], n_neighbors=10, random_state=seed*rnd, n_jobs=-1) for rnd in range(10)])
mi_score=mi_score_arr.mean(axis=0)

df_mi=pd.DataFrame({
                "features": df_feat_transformed.columns, 
                "MI": mi_score
            })
df_mi["rank"]=df_mi["MI"].rank(method="dense",ascending=False).astype(int)
df_mi
```

```python
df_mi_ranked=df_mi.sort_values(by="rank", ascending=True).reset_index(drop=True)
df_mi_ranked
```

```python
features_ranked=df_mi_ranked["features"].to_list()
# features_ranked
```
